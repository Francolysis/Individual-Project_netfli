```python
#read in the csv as dataframe
import pandas as pd
netflix_df = pd.read_csv('netflix_data.csv')

#print (netflix_df.info())
#print (netflix_df.head())

#netflix_df.head()
netflix_df.head()
```

```python
# Subsetting the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']

# Select only the columns of interest
netflix_movies_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]

#printing the first five rows
print (netflix_movies_col_subset.head())
```

```python
# importing matpotlib
import matplotlib.pyplot as plt
import pandas as pd

# Create a figure and increase the figure size
fig_2 = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus year
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], marker='o')

# Create a title
plt.title('Movie Duration by Year of Release')

plt.show
```

```python
# Filter for durations shorter than 60 minutes
short_movies = netflix_movies_col_subset[(netflix_movies_col_subset["duration"] < 60)]
                                         
# Print the first 20 rows of short_movies
print (short_movies.head(20))
```

```python
# Defining an empty list
colors = []

# Iterating over rows of netflix_movies_col_subset
for row, ser in netflix_movies_col_subset.iterrows():
    if ser['genre'] == 'Children':
        colors.append ('red')
    elif ser['genre'] == 'Documentaries':
        colors.append ('blue')
    elif ser ['genre'] == 'Stand-Up':
        colors.append ('green')
    else:
        colors.append ('black')

# Inspecting the first 10 values in your list
print (colors[:10])
```

```python
# Set the figure style and initalize a new figure
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus release_year
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], color=colors, marker='o')

# Create a title and axis labels
plt.title('Movie Duration by Year of Release')
plt.xlabel(['release_year'])
plt.ylabel(netflix_movies_col_subset['duration'])

# Show the plot
plt.show
```
